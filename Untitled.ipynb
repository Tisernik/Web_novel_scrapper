```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Настройка веб-драйвера
service = Service(ChromeDriverManager().install())
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Этот параметр скрывает окно браузера

# Создание экземпляра драйвера
driver = webdriver.Chrome(service=service, options=options)

# Указываем URL страницы
url = "https://novelbin.com/series/star-odyssey#tab-chapters-title"

# Переходим на страницу
driver.get(url)

# Ожидаем некоторое время, чтобы страница успела загрузиться
time.sleep(15)

# Получаем HTML-код страницы
html_content = driver.page_source

# Закрытие драйвера
driver.quit()

# Сохраняем HTML-код в файл
with open("page_source.html", "w", encoding="utf-8") as file:
    file.write(html_content)

print(f"HTML-код страницы успешно сохранён в файл 'page_source.html'.")
```

```python
import requests
from bs4 import BeautifulSoup

# Список кортежей "ссылка-название"
chapters = [
    ('https://novelbin.com/b/star-odyssey/chapter1-lu-yin', 'Chapter 1: Lu Yin'),
    ('https://novelbin.com/b/star-odyssey/chapter2-camp-of-the-seven-sages', 'Chapter 2: Camp of the Seven Sages')
]

def parse_and_save_chapter(chapter_url, chapter_title):
    """Функция для парсинга и сохранения текста главы."""
    try:
        # Скачиваем HTML-документ
        response = requests.get(chapter_url)
        if response.status_code != 200:
            raise Exception(f"Не удалось получить доступ к главе по адресу: {chapter_url}")
        
        # Парсим HTML с помощью BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Находим контейнер с текстом главы
        chapter_text_container = soup.find('div', {'id': 'chr-content'})
        if chapter_text_container is None:
            raise Exception(f"Контент главы не найден по адресу: {chapter_url}")
        
        # Извлекаем текст главы
        chapter_text = chapter_text_container.get_text(strip=True, separator='\n')
        
        # Сохраняем текст главы в файл
        filename = f"{chapter_title.replace(':', '').replace(' ', '_')}.txt"
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(chapter_text)
            
        print(f"Сохранил главу: {filename}")
    except Exception as e:
        print(f"Ошибка при обработке главы: {e}")

# Обрабатываем каждую главу
for url, title in chapters:
    parse_and_save_chapter(url, title)
```
